```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread, imsave

orig_img = imread('01.jpg')
plt.imshow(orig_img)
plt.show()
```

```python
test1 = orig_img.copy()

y_min, y_max = 90, 220
x_min, x_max = 160, 320

h=10

area_width = (x_max - x_min) // h
area_height = (y_max - y_min) // h

average_colors = np.zeros((h, h, 3))

for i in range(h):
    for j in range(h):
        x1 = x_min + i * area_width
        y1 = y_min + j * area_height
        x2 = x1 + area_width
        y2 = y1 + area_height

        region = test1[y1:y2, x1:x2]

        average_color = np.mean(region, axis=(0, 1))

        test1[y1:y2, x1:x2] = average_color

plt.imshow(test1)
plt.axis('off')
plt.show()
```
